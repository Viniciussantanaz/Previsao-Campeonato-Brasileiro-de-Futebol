# Previsão do Campeonato Brasileiro com Métodos estatísicos.

### O objetio é calcular a probabilidade de vitória do time A ou time B de vencer a partida usando probabilidade binomial.

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$


#### - P = (Média.FinaLizações certas timea A /  Média.Finalizações tiem A) * (Média.Gols cedidos time B / Média.Finalizações cedidas time B)
#### - n = Média.Finalizações tiem A 
#### - k = 1...6


```python
import pandas as pd
from scipy.stats import binom
import numpy as np
```

### Importando as tabelas

```python
tabelas = pd.read_html('https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2024_-_S%C3%A9rie_A') # Importando as teblas de jogos e de classificação
```

```python
tabela_jogos = tabelas[7]
tabela_campeonato = tabelas[6]
tabela_campeonato.replace({'Botafogo (X)':'Botafogo', 'Palmeiras (X)':'Palmeiras'}, inplace=True) #Ajustando o nome 'Botafogo (X)' e ''Palmeiras (X)'
tabela_campeonato
```

```python
tabela_jogos.rename(columns={'Casa \ Fora':'Casa_Fora'}, inplace=True) #Trocando o nome para evitar a mensagem abaixo
tabela_jogos.set_index('Casa_Fora',inplace=True)
tabela_jogos
```

### Tratando a tabela de jogos

```python
# Dicionário com a chave sendo os apelidos e valor o nome dos tiems.
nomes_times = list(tabela_jogos.index)
apelido_times = list(tabela_jogos.columns)
aplido_para_nomes = dict(zip(apelido_times,nomes_times))
tabela_jogos = tabela_jogos.reset_index()
aplido_para_nomes
```

```python
#Tratando a tabela de jogos
tabela_jogos.set_index('Casa_Fora', inplace=True)
tabela_jogos = tabela_jogos.unstack().reset_index() 
tabela_jogos.rename(columns={'level_0':'time_fora', 'Casa_Fora':'time_casa', 0:'Placar'}, inplace=True)
tabela_jogos.set_index('time_fora', inplace=True)
tabela_jogos
```

```python
tabela_jogos.rename(index=aplido_para_nomes, inplace=True) #Usando o dicionário para deixar os nomes padrõs
tabela_jogos.reset_index(inplace=True)
tabela_jogos
```

```python
tabela_jogos = tabela_jogos[tabela_jogos.Placar != '—'] #Excluido as linhas em que os nome time casa e time fora são iguais.
tabela_jogos = tabela_jogos.fillna('A jogar')
jogos_realizados = tabela_jogos[tabela_jogos.Placar.str.contains('–')] #Jogos realizados
jogos_faltantes = tabela_jogos[~tabela_jogos.Placar.str.contains('–')] #Jogos que faltam 
jogos_faltantes.reset_index(drop=True, inplace=True)
jogos_faltantes = jogos_faltantes.copy() # Para evitar a ambiguidade entre trabalhar com uma visão ou uma cópia do DataFrame original.
jogos_faltantes.drop(columns='Placar', inplace=True)
jogos_faltantes
```

```python
jogos_realizados = jogos_realizados.copy() # Para evitar a ambiguidade entre trabalhar com uma visão ou uma cópia do DataFrame original.
jogos_realizados[['Gols_casa', 'Gols_fora']] = jogos_realizados['Placar'].str.split('–', expand=True).astype(int) 
jogos_realizados.drop('Placar', axis=1, inplace=True)
jogos_realizados.reset_index(drop=True, inplace=True)
jogos_realizados
```

```python
#Importando a tabela de finalizações
df_stats = pd.read_excel('stats.xlsx')
df_stats.replace({'RB Bragantino':'Red Bull Bragantino',
                  'Atlético-MG':'Atlético Mineiro',
                  'Atlético-GO':'Atlético Goianiense'}, inplace=True) #Ajustando o nome de alguns times da tabela stats para ficar padrão.
df_stats #Tabale de finalizações
```

### Importando a tabela de finalizações

```python
tabela_estatistica = pd.merge(jogos_realizados, df_stats, on=['time_casa', 'time_fora'], how='inner') 
tabela_estatistica[tabela_estatistica.time_fora == 'São Paulo']
tabela_estatistica #Tabela de estatíscas
```

```python
tabela_medias = {'time':[], 
                     'media_gol_cedido_fora':[],
                     'media_gol_cedido_casa':[],
                     'media_finalizacao_em_casa':[],
                     'media_finalizacao_fora_de_casa':[],
                     'media_finalizacao_certa_em_casa':[],
                     'media_finalizacao_certa_fora_de_casa':[],
                     'media_finalizacoes_cedidas_casa':[],
                     'media_finalizacoes_cedidas_fora':[]}
for time in nomes_times:
    tabela_medias['time'].append(time)
    tabela_medias['media_gol_cedido_fora'].append(tabela_estatistica[tabela_estatistica.time_fora == time].Gols_casa.mean())
    tabela_medias['media_gol_cedido_casa'].append(tabela_estatistica[tabela_estatistica.time_casa == time].Gols_fora.mean())
    tabela_medias['media_finalizacao_em_casa'].append(tabela_estatistica[tabela_estatistica.time_casa == time].chutes_gol_casa.mean())
    tabela_medias['media_finalizacao_fora_de_casa'].append(tabela_estatistica[tabela_estatistica.time_fora == time].chutes_fora.mean())
    tabela_medias['media_finalizacao_certa_em_casa'].append(tabela_estatistica[tabela_estatistica.time_casa == time].Gols_casa.mean())
    tabela_medias['media_finalizacao_certa_fora_de_casa'].append(tabela_estatistica[tabela_estatistica.time_casa == time].chutes_gol_fora.mean())
    tabela_medias['media_finalizacoes_cedidas_casa'].append(tabela_estatistica[tabela_estatistica.time_casa == time].Gols_fora.mean())
    tabela_medias['media_finalizacoes_cedidas_fora'].append(tabela_estatistica[tabela_estatistica.time_fora == time].Gols_casa.mean())

tabela_medias = pd.DataFrame(tabela_medias)

tabela_medias #Tabela com as médias das métricas.

```

# Calculandno a probabilide dos jogos

```python
# n = num.Finaçizacoes
# k = [1,2,3,4,5,6]
# p = (media.Finaçizacoes certas timea A /  media.Finaçizacoes tiem A) * (media.Gols cedidos time B / media.finalizacoes cedidas time B)
# q = 1-p
# probabilidade = comb(n,k)*(p**k)*(q**(n-k)
```

```python
P_time_casa = []
P_time_fora = []
P_empate = []
for time in jogos_faltantes.index:
    time_casa = jogos_faltantes.iloc[time,1]
    time_fora = jogos_faltantes.iloc[time,0]
    
    # Probabilidade do Time da casa ganhar.

    finalizacao_time_casa = tabela_medias.loc[tabela_medias.time == time_casa, 'media_finalizacao_em_casa'].values[0]
    finalizacao_certa_time_casa = tabela_medias.loc[tabela_medias.time == time_casa, 'media_finalizacao_certa_em_casa'].values[0]
    gols_sofrido_time_fora = tabela_medias.loc[tabela_medias.time == time_fora, 'media_gol_cedido_fora'].values[0]
    finalizacao_cedida_time_fora = tabela_medias.loc[tabela_medias.time == time_fora, 'media_finalizacoes_cedidas_fora'].values[0]
    
    n = finalizacao_time_casa.astype(int)
    p = (finalizacao_certa_time_casa / finalizacao_time_casa) * (gols_sofrido_time_fora / finalizacao_cedida_time_fora)
    k = np.arange(1,7)
    
    if p > 1:
        probabilidade_time_casa = 0.99 # Confronto Fortaleza X Vasco, Vasco teve probabilidade de mais 100% de marcar gol em todas as chances.
                                       # para resolver coloquei 99% de chance de vitória para evitar Erro.
    else:
         probabilidade_time_casa = (binom.pmf(k,n,p))
    
    
    
    #Probabilidade do time de fora ganhar
    
    finalizacao_time_fora = tabela_medias.loc[tabela_medias.time == time_fora, 'media_finalizacao_fora_de_casa'].values[0]
    finalizacao_certa_time_fora = tabela_medias.loc[tabela_medias.time == time_fora, 'media_finalizacao_certa_fora_de_casa'].values[0]
    gols_sofrido_time_casa = tabela_medias.loc[tabela_medias.time == time_casa, 'media_gol_cedido_casa'].values[0]
    finalizacao_cedida_time_casa =  tabela_medias.loc[tabela_medias.time == time_casa, 'media_finalizacoes_cedidas_casa'].values[0]

    n = finalizacao_time_fora.astype(int)
    p = (finalizacao_certa_time_fora / finalizacao_time_fora) * (gols_sofrido_time_casa / finalizacao_cedida_time_casa)
    k = np.arange(1,7)

    probabilidade_time_fora = binom.pmf(k,n,p)

    #Probabilidade de empate
    probalidade_empate = sum([probabilidade_time_fora[p] *  probabilidade_time_casa[p] for p in range(6)])

    # Condições 
    if (probalidade_empate > sum(probabilidade_time_casa) and probalidade_empate > sum(probabilidade_time_fora)) or (sum(probabilidade_time_fora) == sum(probabilidade_time_casa)):
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'Pts'] += 1
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'Pts'] += 1
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'E'] += 1
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'E'] += 1
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'J'] += 1
        tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'J'] += 1


    
    elif sum(probabilidade_time_casa) > sum(probabilidade_time_fora):
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'Pts'] += 3
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'V'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'D'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'J'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'J'] += 1


    else:
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'Pts'] += 3
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'D'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'V'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_casa, 'J'] += 1
         tabela_campeonato.loc[tabela_campeonato.Equipevde == time_fora, 'J'] += 1

    P_time_casa.append(sum(probabilidade_time_casa))
    P_time_fora.append(sum(probabilidade_time_fora))
    P_empate.append(probalidade_empate)
```

### Resultados

```python
jogos_faltantes = jogos_faltantes.assign(p_time_fora=P_time_fora, p_time_casa=P_time_casa, p_empate=P_empate )
jogos_faltantes
```
